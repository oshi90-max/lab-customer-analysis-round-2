```python
def clean(file):                                                             #Cleaning data function 
    file = file.drop(['Unnamed: 0'], axis=1)
    file['State'].value_counts(dropna=False)
    file['Vehicle Type'] = file['Vehicle Type'].fillna('not A')
    file['Vehicle Type'].value_counts(dropna=False)
    file = file[file['State'].isna()==False]
    file = file[file['Months Since Last Claim'].isna()==False] 
    file = file[file['Vehicle Class'].isna()==False] 
    file['Effective To Date'] = pd.to_datetime(file['Effective To Date']) 
    file['Effective To Date Month'] = file['Effective To Date'].dt.month 
    file[file['Effective To Date Month'].isin([1, 2, 3])] #12
    columns = [] 
    for i in file.columns:
        columns.append(i.lower())
    file.columns = columns
    return file
```

```python
# We use the describe values to create a function which will help put the row values in 'total claim amount' to different bins
def c_bins(row):
    if row < 272:
        return ' < 25%'
    elif row < 383:
        return '25%-50%'
    elif row < 548:
        return '50%-75%'
    else:
        return '> 75%'
# Go and get every row of 'total claim amount' and apply the bins function to every row
file['claim bin'] = file['total claim amount'].apply(lambda row: c_bins(row))   # Adding the new row to the dataframe using the c_bins function 
file.head()
```
